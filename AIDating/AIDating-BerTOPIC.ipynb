数据预处理，去掉@人的部分

```python
#preprocessing
import re
#去掉@别人:从遇到一个@开始到下一个空格之前的部分删掉
def remove_AT(text):
    if not isinstance(text, str):
        return text
    # 只删除 "@id" 部分,保留评论内容
    pattern = r'@[^\s]+'
    return re.sub(pattern+'', '', text)

print(df2['content'][8])
print(remove_AT(df2['content'][8]))

#把叠词合并
def replace_excessive_ha(text):
    if not isinstance(text, str):
        return text

    # 替换 3 个及以上的 'ha' 为 '哈哈哈'
    pattern = r'哈{3,}'
    return re.sub(pattern, lambda x: '哈哈哈', text)

def replace_excessive_a(text):
    if not isinstance(text, str):
        return text

    # 替换 3 个及以上的 'ha' 为 '哈哈哈'
    pattern = r'啊{3,}'
    return re.sub(pattern, lambda x: '啊啊啊', text)

#以空格为开头的数据，把所有空格的部分都删掉
def remove_space(text):
    if not isinstance(text, str):return text
    return text.strip()
dfClean = pd.DataFrame()
dfClean['content'] = df2['content'].apply(remove_AT)
dfClean['content'] = dfClean['content'].apply(replace_excessive_ha)
dfClean['content'] = dfClean['content'].apply(replace_excessive_a)
dfClean['content'] = dfClean['content'].apply(remove_space)
dfClean = dfClean[dfClean.apply(lambda row: all(len(str(x)) >= 2 for x in row), axis=1)]




try:
    with pd.ExcelWriter('data/xhs/'+dataID+'/output-660b3ac5000000001a0110a8.xlsx', engine='openpyxl', mode='a') as writer:
        dfClean[['content']].to_excel(writer, sheet_name='RmAT', index=False)
except FileNotFoundError:
    with pd.ExcelWriter('data/xhs/'+dataID+'/output-660b3ac5000000001a0110a8.xlsx', engine='openpyxl', mode='w') as writer:
        dfClean[['content']].to_excel(writer, sheet_name='RmAT', index=False)
```

```python
DataCleaned=pd.read_excel('data/xhs/'+dataID+'/output-660b3ac5000000001a0110a8.xlsx',sheet_name='RmAT')
sentences=DataCleaned['content'].tolist()
print('文本条数: ', len(sentences))
print('预览第一条: ', sentences[0])
```

```python
vertorizer_model=None
```

```python
model_name="bert-base-chinese"
# model_name="hfl/chinese-bert-wwm"
model=BertModel.from_pretrained(model_name)
tokenizer=BertTokenizer.from_pretrained(model_name)
```

将数据每16条一组分批

```python
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()
batch_size=16
data_loader=DataLoader(sentences,batch_size=batch_size, shuffle=False)
for batch in data_loader:
    print(len(batch),batch)
```

文本转向量



```python
cls_embeddings = []

# 使用tqdm显示处理进度
for batch_sentences in tqdm(data_loader):
    #通过tokenizer分词，并且padding补齐，添加CLS符号
    inputs = tokenizer(batch_sentences, padding=True, truncation=True, return_tensors="pt", max_length=512)
    inputs.to(device)
    #文本转向量
    with torch.no_grad():#不调参，不对模型进行更新，只有前向传播 
        outputs = model(**inputs)
    cls_embeddings.append(outputs.last_hidden_state[:, 0].cpu().numpy()) # 只取CLS对应的向量#取出每一个句子CLS对应的向量
cls_embeddings_np = np.vstack(cls_embeddings)#纵向堆叠
output_file = "emb.npy"
np.save(output_file, cls_embeddings_np)
embeddings = np.load(output_file)
```

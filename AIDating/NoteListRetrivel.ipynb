获取每一个找到的note的id

**GLOW**

```python
# 导入所需库
import re

# 定义函数来检测关键词并获取对应结果
def extract_note_ids(file_path):
    note_ids = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            # 检查是否包含关键词'note_id'
            if 'note_id' in line:
                # 使用正则表达式提取note_id的值
                match = re.search(r'note_id:\s*(\S+)', line)
                if match:
                    note_ids.append(match.group(1))  # 添加到结果列表
    return note_ids

# 使用示例
file_path = 'List/AIGirlfriend.txt'  # 替换为你的txt文件路径
note_ids = extract_note_ids(file_path)

# 输出结果
print("提取的note_id:", note_ids)
print(len(note_ids))
output_file_path = 'List/AIGirlfriend_note_ids_list.txt'
with open(output_file_path, 'w', encoding='utf-8') as file:
    for note_id in note_ids:
        file.write(f"{note_id}\n")  # 每个note_id一行
```

```python
file_path = 'List/AIGirlfriend_note_ids_list.txt' 
# 初始化一个空列表
lines_list = []

# 打开文件并读取内容
with open(file_path, 'r', encoding='utf-8') as file:
    lines_list = [line.strip() for line in file.readlines()]

# 输出结果
print(lines_list)
```
